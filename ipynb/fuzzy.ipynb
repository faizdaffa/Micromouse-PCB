```python
import numpy as np
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)
import matplotlib.pyplot as plt
```

```python
xDekat = [0,10,20,40]
yDekat = [1,1,0,0]
xJauh = [0,20,30,40]
yJauh = [0,0,1,1]
xSedang = [0,10,20,30,40]
ySedang = [0,0,1,0,0]

fig, ax1 = plt.subplots(figsize=(10, 4))
ax1.set_title('Fungsi Keanggotaan Input (Ultrasonik Kiri, Depan, Kanan)', weight='bold')
ax1.set_xlabel('Jarak (cm)', weight='bold')
ax1.set_ylabel('Membership Value', weight='bold')

ax1.text(2.5  ,1.05, 'Dekat', weight='bold')
ax1.text(18.5 ,1.05, 'Sedang', weight='bold')
ax1.text(33.5 ,1.05, 'Jauh', weight='bold')
ax1.set_ylim([0,1.15])

ax1.set_facecolor('#ffff99')

ax1.plot(xSedang,ySedang, color='g')
ax1.plot(xDekat,yDekat, color='b')
ax1.plot(xJauh,yJauh, color='r')
```

```python
xKiri = [-90,-45,0,90]
yKiri = [1,1,0,0]
xKanan = [-90,0,45,90]
yKanan = [0,0,1,1]
xLurus = [-90,-45,0,45,90]
yLurus = [0,0,1,0,0]

fig, ax = plt.subplots(figsize=(10, 4))

ax.set_title('Fungsi Keanggotaan Sudut', weight='bold')
ax.set_xlabel('Sudut (derajat)', weight='bold')
ax.set_ylabel('Membership Value', weight='bold')

ax.text(-73  ,1.05, 'Kiri', weight='bold')
ax.text(-6 ,1.05, 'Lurus', weight='bold')
ax.text(60 ,1.05, 'Kanan', weight='bold')
ax.set_xlim([-100,100])
ax.set_ylim([0,1.15])

#ax.set_yticklabels(ax.get_yticks(), weight='bold')
ax.set_xticklabels(ax.get_xticks(), weight='bold')

ax.xaxis.set_major_locator(MultipleLocator(45))
ax.xaxis.set_major_formatter('{x:.0f}')

ax.set_facecolor('#cce6ff')

ax.plot(xLurus,yLurus, color='g')
ax.plot(xKiri,yKiri, color='b')
ax.plot(xKanan,yKanan, color='r')
```
